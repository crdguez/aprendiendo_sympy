```python
from sympy import init_session, lambdify, S, symbols, pi
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline



def tabla_representacion(expr, x_min=-10., x_max=10., num_puntos=20):
    # Parámetros:
    #    x_min: valor mínimo de la x (si no ponemos nada toma -10 por defecto)
    #    x_max: valor mínimo de la x (+10 por defecto)
    #    num_puntos: número de puntos de la gráfica (20 puntos por defecto)
    # Ejemlos de uso:
    #    tabla_representacion('x**2')
    #    tabla_representacion('1/x', -8 , 8, 40)
    #    tabla_representacion('cos(x)',-6.3, 6.3, 30)
    
    exp = S(expr)
    x = symbols('x')

    f = lambdify(x, exp, 'numpy')
    x = np.linspace(x_min,x_max,num_puntos)
    plt.plot(x,f(x),'o')
    plt.grid(True)
    plt.axis('equal')
    plt.show()

    display(pd.DataFrame({'x':x,'y':f(x)}))
    



```

```python
tabla_representacion('x**2')
```

```python
tabla_representacion('cos(x)',-6.3, 6.3, 30)
```

```python
tabla_representacion('1/x', -8 , 8, 40)
```
